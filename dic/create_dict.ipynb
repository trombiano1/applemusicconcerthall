```python
import json
dic = {}
d = []

with open('all.txt') as f:
    s = f.read()
    d.append(json.loads(s))
```

```python
last = 'z'
for d_ in d:
    for composer in d_['composers']:
        # print(composer['name'])
        if ord(composer['complete_name'].split(" ")[-1][0].upper()) - ord(last) == 1 or last == 'z':
            # print(ord(last) - ord(composer['complete_name'].split(" ")[-1][0].upper()))
            last = composer['complete_name'].split(" ")[-1][0].upper()
            print("<option value='"+last+"' disabled>"+last+"</option>")
        if composer['complete_name'].split(" ")[-1][0].upper() == 'R' and composer['complete_name'].split(" ")[-1][0].upper() != last:
            # print(ord(last) - ord(composer['complete_name'].split(" ")[-1][0].upper()))
            last = composer['complete_name'].split(" ")[-1][0].upper()
            print("<option value='"+last+"' disabled>"+last+"</option>")
        if composer['complete_name'].split(" ")[-1][0].upper() == 'V' and composer['complete_name'].split(" ")[-1][0].upper() != last:
            # print(ord(last) - ord(composer['complete_name'].split(" ")[-1][0].upper()))
            last = composer['complete_name'].split(" ")[-1][0].upper()
            print("<option value='"+last+"' disabled>"+last+"</option>")
        print("<option value='"+composer['complete_name']+"'>"+composer['complete_name']+"</option>")
        dic[composer['complete_name']] = int(composer['id'])
```
