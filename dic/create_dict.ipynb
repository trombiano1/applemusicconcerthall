```python
import json
dic = {}
d = []

# alp = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']

alp = ['all']

for a in alp:
    with open(a+'.txt') as f:
        s = f.read()
        d.append(json.loads(s))
```

```python
for d_ in d:
    for composer in d_['composers']:
        # print(composer['name'])
        print("<option value='"+composer['complete_name']+"'>"+composer['name']+"</option>")
        dic[composer['complete_name']] = int(composer['id'])
```
